```python
x = pd.read_html("https://afltables.com/afl/teams/stkilda/allgames.html")
```

```python
def only_results(df):
    size = len(df)
    return df.head(size-2)
```

```python
def slim_table(df, cols=["F", "A"]):
    return df[cols]
```

```python
only_results(slim_table(x[13]))
```

```python
# 2004 - 2008

```

```python
def calculate_percentage(df):
    df["percentage"] = (df["F"] / df["A"]) * 100
    return df
```

```python
calculate_percentage(only_results(slim_table(x[13])))
```

```python
def clean_table(year):
    df = (only_results(slim_table(x[2022-year])))
    df.columns = ["F", "A"]
    return df
```

```python
all_years = []
for year in range(2011, 2022):
    all_years.append(clean_table(year))

vertical_stack = pd.concat(all_years, axis=0).reset_index()
```

```python
vertical_stack["rolling_sum_points_scored"] = vertical_stack["F"].rolling(window=30).sum()
vertical_stack["rolling_sum_points_against"] = vertical_stack["A"].rolling(window=30).sum()
```

```python
vertical_stack["rolling_sum_points_differential"] = vertical_stack["rolling_sum_points_scored"] - vertical_stack["rolling_sum_points_against"]
```
